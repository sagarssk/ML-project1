### __dummy encoding___

```python
#dataframe["gender"]=dataframe.gender.astype(float)
dataframe.dtypes
```

```python
dataframe["Dependents"]=dataframe["Dependents"].map({'Yes':1,'No':0})


```

```python
dataframe["Partner"]=dataframe["Partner"].map({'Yes':1,'No':0})
```

```python
dataframe["gender"]=dataframe["gender"].map({'Female':0,'Male':1})
```

```python
dataframe["PhoneService"]=dataframe["PhoneService"].map({'Yes':1,'No':0})
```

```python
dataframe["PaperlessBilling"]=dataframe["PaperlessBilling"].map({'Yes':1,'No':0})

```

```python
dataframe["Churn"]=dataframe["Churn"].map({'Yes':1,'No':0})

```

```python
dataframe=pd.get_dummies(dataframe,columns=["PaymentMethod"],drop_first=True)
```

```python
dataframe=pd.get_dummies(dataframe,columns=["MultipleLines","InternetService","OnlineSecurity"])
```

```python
dataframe=pd.get_dummies(dataframe,columns=['OnlineBackup','DeviceProtection','TechSupport'])
```

```python
dataframe=pd.get_dummies(dataframe,columns=['StreamingTV','StreamingMovies','Contract'])
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
dataframe[['customerID']]=lb.fit_transform(dataframe[["customerID"]])
```

```python
dataframe["customerID"].head()
```

```python
x=dataframe.drop("Churn",axis=1)
x.head()
```

### Recursive feature elimination 


```python
from sklearn.feature_selection import RFE
```

```python
lr2=LinearRegression()
```

```python
xy=RFE(lr2,35)
```

```python
xy.fit(x_train,y_train)
```

```python
xy.support_
```

```python
xy.ranking_
```

```python
x_train.columns
```

```python
s=x_train.columns[xy.support_]
```

```python
lr2.fit(x_train[s],y_train)
```

```python
lr2.predict(x_train[s])
```

### measuring model accuracy

```python
from sklearn.metrics import r2_score
```

```python
Y_train=lr2.predict(x_train[s])
```

```python
r2_score(Y_train,y_train)
```
